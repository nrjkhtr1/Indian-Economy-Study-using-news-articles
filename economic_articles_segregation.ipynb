Here I am using some economics dictionary to extracted from above links to keep count of words present in the articles. This step can help us to remove all the documents which are not related to economy.

#### First dicitonary

```python
def lowertext(words):
    words = words[0:-1]
    return words.lower()

f = open("eco_dictionary1.txt", "r")
text = f.readlines()
f.close()

eco_dictionary1 = []
for words in text:
    eco_dictionary1.append(lowertext(words))

len(eco_dictionary1)
```

```python
eco_dictionary1[1]
```

#### Second dicitonary

```python
f = open("eco_dictionary2.txt", "r")
text = f.readlines()
f.close()

eco_dictionary2 = []
for words in text:
    eco_dictionary2.append(lowertext(words))

eco_dictionary2 = list(set(eco_dictionary2))
len(eco_dictionary2)
```

```python
eco_dictionary2[1]
```

#### Merging both the dictonaries

```python
eco_dictionary = eco_dictionary1 + eco_dictionary2
eco_dictionary = set(list(eco_dictionary))

len(eco_dictionary)
```

```python
import pandas as pd
df = pd.read_csv("supervised_text.csv", index_col = None)

df = df.drop(["Unnamed: 0"], axis = 1)
df.head()
```

```python
df['Text'] = df['Text'].str.lower()
df.head()
```

I am defining two new columns <b>articles_contain_words</b> and <b>count_eco_words</b>.

<b>count_eco_words</b> = how many eco-words are present in the article

<b>articles_contain_words</b> = how many unique eco-words are present in the article


Example sentence: ["free-trade zone", "the", "coal", "capital consumption", "free-trade zone"]<br>
<b>count_eco_words</b> = 3 ("free-trade zone", "capital consumption", "free-trade zone")<br>
<b>articles_contain_words</b> = 2 ("free-trade zone", "capital consumption")

```python
df['count_eco_words'] = 0

# eco = ["price", "value", "stock", "union"]

for string in eco_dictionary:
#     print(string)
    df['count_eco_words'] += df['Text'].str.count(string)
    
df.head()
```

```python
df_eco = df[df['count_eco_words'] != 0]
df_non_eco = df[df['count_eco_words'] == 0]
```

```python
df_non_eco.tail()
```

```python
df_eco.head()
```

```python
df['articles_contain_words'] = 0

for string in eco_dictionary:
    bool_list = df['Text'].str.contains(string)
    bool_list = [int(elem) for elem in bool_list]
    df['articles_contain_words'] += bool_list
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
plt.plot(df["articles_contain_words"].to_list())
```

```python
plt.figure(figsize=(15,8))
plt.plot(df["count_eco_words"].to_list())
```

Now, we have to decide a threshold on the eco-words to accept it as economic article.

```python
df["eco_words_bool"] = 0
df["eco_words_bool_count"] = 0

df ["eco_words_bool"] = df.apply(lambda x: x.articles_contain_words > 20, axis=1)
df ["eco_words_bool_count"] = df.apply(lambda x: x.count_eco_words > 30, axis=1)
```
